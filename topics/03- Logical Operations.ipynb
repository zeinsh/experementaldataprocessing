```python
import numpy as np
% matplotlib notebook
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
```

```python
import cv2 as cv
img = cv.imread('./data/MODEL.jpg',0)
thresh_img=(img>=200)
```

```python
not_img=np.bitwise_not(thresh_img)
plt.figure()
plt.subplot(1,2,1)
plt.title('Original Image')
imshow(thresh_img,cmap='gray')
plt.subplot(1,2,2)
plt.title('Not operation')
imshow(not_img,cmap='gray')
```

```python
def getCoordinates(num):
    return num//400,num%400
def dist(x1,y1,x2,y2):
    return np.sqrt((x1-x2)**2+(y1-y2)**2)
def inCircle(num):
    x1=150
    y1=150
    r=75
    x2,y2=getCoordinates(num)
    return dist(x1,y1,x2,y2)<=r

circle=inCircle(np.arange(300*400).reshape(300,400))
plt.figure()
plt.title('binary circle')
imshow(circle,cmap='gray')
```

```python
or_img=circle+thresh_img
and_img=circle*thresh_img
xor_img=np.bitwise_xor(circle,thresh_img)
plt.figure()
plt.subplot(2,3,1)
plt.title('Original')
imshow(thresh_img,cmap='gray')
plt.subplot(2,3,2)
plt.title('Circle')
imshow(circle,cmap='gray')
plt.subplot(2,3,3)
plt.title('Not operation')
imshow(not_img,cmap='gray')
plt.subplot(2,3,4)
plt.title('Or operation')
imshow(or_img,cmap='gray')
plt.subplot(2,3,5)
plt.title('And operation')
imshow(and_img,cmap='gray')
plt.subplot(2,3,6)
plt.title('Xor operation')
imshow(xor_img,cmap='gray')
```
