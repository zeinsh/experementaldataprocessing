```python
import numpy as np
% matplotlib notebook
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
```

```python
import cv2 as cv
img = cv.imread('./data/stones.jpg',0)
```

```python
img1=img+10
img2=img-20
img3=img*1.3
img4=img*0.7
img5=img**1.2

plt.figure()
plt.subplot(2,3,1)
plt.title('Original')
imshow(img,cmap='gray')
plt.subplot(2,3,2)
plt.title('increase brightness')
imshow(img1,cmap='gray')
plt.subplot(2,3,3)
plt.title('decrease brightness')
imshow(img2,cmap='gray')
plt.subplot(2,3,4)
plt.title('increase contrast')
imshow(img3,cmap='gray')
plt.subplot(2,3,5)
plt.title('decrease contrast')
imshow(img4,cmap='gray')
plt.subplot(2,3,6)
plt.title('Original^1.2 ')
imshow(img5,cmap='gray')
plt.tight_layout()
```
