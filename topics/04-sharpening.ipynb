```python
import numpy as np
% matplotlib notebook
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
```

```python
import cv2 as cv
img = cv.imread('./data/stones.jpg',0)
plt.figure()
plt.title('Original')
imshow(img,cmap='gray')
```

```python
from scipy import ndimage
# Lablasian Filters
lablas=np.array([[-1,-1,-1],
                 [-1, 8,-1],
                 [-1,-1,-1]])
filtered=ndimage.convolve(img,lablas)
plt.figure()
plt.subplot(1,2,1)
plt.title('After filtering')
imshow(filtered,cmap='gray')
filtered=255*(filtered-filtered.min())/filtered.max()
plt.subplot(1,2,2)
plt.title('After filtering and normalization')
imshow(filtered,cmap='gray')
```

```python
sharpened = img + filtered
sharpened = 255*((sharpened-sharpened.min())/sharpened.max())
imshow(sharpened,cmap='gray')
```
