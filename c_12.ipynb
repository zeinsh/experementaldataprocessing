```python
import soundfile as sf
import matplotlib.pyplot as plt
import numpy as np
from plotmd import plotVector
from helpersmd import fourierTransform
```

```python
data, fs = sf.read('roma.wav')
N=data.shape[0]
Cn=np.real(np.fft.fft(data)[:N/2])
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plotVector(data,"Ro-ma waveform ","n",'f(n)')
plt.subplot(1,2,2)
plotVector(Cn,"Ro-ma spectrum","n",'F(n)')
plt.show()
```
