# Выделение контуров частотными фильтрами

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
from lib.helpersmd import readFromDat
%matplotlib inline
```

```python
def getHistogram(img,mx):
    return np.array([np.sum(img==i) for i in range(mx+1)])
```

## Explore image and histogram

```python
from PIL import Image
model = np.array(Image.open("./data/MODEL.jpg"))
size=model.shape
imshow(model,cmap='gray')
```

```python
hist=getHistogram(model,255)
plt.bar(np.arange(256),hist)
```

## 2. Threshold transform

```python
threshimg=np.zeros_like(model)
w=model>200
b=model<200
threshimg[w],threshimg[b]=255,0
```

```python
plt.figure()
imshow(threshimg,cmap='gray')
```

## 3. Apply low pass filter 

- I used in this example Gaussian Filter
- You can use LPF builded in previous parts of the course

```python
from scipy.ndimage.filters import gaussian_filter
gfltr=gaussian_filter(threshimg,0.5)
imshow(gfltr,cmap='gray')
```

## 4. Get the contour 

Countor=thresholdImage-GaussianImage

```python
im=gfltr-threshimg
imshow(im,cmap='gray')
```
